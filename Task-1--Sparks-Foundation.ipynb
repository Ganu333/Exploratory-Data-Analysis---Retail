**Visualization**

```python
a=pd.DataFrame(df.groupby("Ship Mode")["Sales"].sum().sort_values(ascending=False))
a.reset_index(inplace=True)
a.columns=["Ship Mode","Sales"]


b=pd.DataFrame(df.groupby("Ship Mode")["Profit"].sum().sort_values(ascending=False))
b.reset_index(inplace=True)
b.columns=["Ship Mode","Profit"]



fig_1= make_subplots(rows=1,cols=2,subplot_titles=("Ship Mode Vs Sales","Ship Mode Vs Profit"))
fig_1.add_trace(go.Bar(x=a["Ship Mode"],y=a["Sales"],marker=dict(color=[6,7,8,9])),1,1)
fig_1.add_trace(go.Bar(x=b["Ship Mode"],y=b["Profit"],marker=dict(color=[6,7,8,9])),1,2)

fig_1.update_xaxes(title_text="Ship Mode",row=1,col=1)
fig_1.update_xaxes(title_text="Ship Mode",row=1,col=2)
fig_1.update_yaxes(title_text="Sales",row=1,col=1)
fig_1.update_yaxes(title_text="Profit",row=1,col=2)

fig_1.update_layout(showlegend=False)
```

```python
c=pd.DataFrame(df.groupby("Segment")["Sales"].sum().sort_values(ascending=False))
c.reset_index(inplace=True)
c.columns=["Segment","Sales"]


d=pd.DataFrame(df.groupby("Segment")["Profit"].sum().sort_values(ascending=False))
d.reset_index(inplace=True)
d.columns=["Segment","Profit"]



fig_2= make_subplots(rows=1,cols=2,subplot_titles=("Segment Vs Sales","Segment Vs Profit"))
fig_2.add_trace(go.Bar(x=c["Segment"],y=c["Sales"],marker=dict(color=[4,5,6])),1,1)
fig_2.add_trace(go.Bar(x=d["Segment"],y=d["Profit"],marker=dict(color=[1,2,3])),1,2)

fig_2.update_xaxes(title_text="Segment",row=1,col=1)
fig_2.update_xaxes(title_text="Segment",row=1,col=2)
fig_2.update_yaxes(title_text="Sales",row=1,col=1)
fig_2.update_yaxes(title_text="Profit",row=1,col=2)

fig_2.update_layout(showlegend=False)
```

```python
c=pd.DataFrame(df.groupby("Region")["Sales"].sum().sort_values(ascending=False))
c.reset_index(inplace=True)
c.columns=["Region","Sales"]


d=pd.DataFrame(df.groupby("Region")["Profit"].sum().sort_values(ascending=False))
d.reset_index(inplace=True)
d.columns=["Region","Profit"]



fig_3= make_subplots(rows=1,cols=2,subplot_titles=("Region Vs Sales","Region Vs Profit"))
fig_3.add_trace(go.Bar(x=c["Region"],y=c["Sales"],marker=dict(color=[4,5,6,7])),1,1)
fig_3.add_trace(go.Bar(x=d["Region"],y=d["Profit"],marker=dict(color=[4,5,6,7])),1,2)

fig_3.update_xaxes(title_text="Region",row=1,col=1)
fig_3.update_xaxes(title_text="Region",row=1,col=2)
fig_3.update_yaxes(title_text="Sales",row=1,col=1)
fig_3.update_yaxes(title_text="Profit",row=1,col=2)

fig_3.update_layout(showlegend=False)
```

```python
state_sales=pd.DataFrame(df.groupby("State")["Sales"].sum().sort_values(ascending=False))
state_sales.reset_index(inplace=True)
state_sales.columns=['State',"Sales"]

fig_7=px.bar(state_sales,x="State",y="Sales",title="State Vs Sales",
            labels={"":"Sales"},color_discrete_sequence=px.colors.sequential.RdBu)
fig_7
```

```python
state_profit=pd.DataFrame(df.groupby("State")["Profit"].sum().sort_values(ascending=False))
state_profit.reset_index(inplace=True)
state_profit.columns=['State',"Profit"]

fig_7=px.bar(state_profit,x="State",y="Profit",title="State Vs Profit",
            labels={"":"Profit"},color_discrete_sequence=px.colors.sequential.RdBu)
fig_7

```
